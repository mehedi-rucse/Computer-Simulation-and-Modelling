Bezier Curve

```python
import pygame as pg

pg.init()
pg.display.set_caption("Bezier Curve")

h,w = 1400,750
x,y = 400,400
screenSize = (h,w)
screen = pg.display.set_mode(screenSize)

f = pg.font.get_fonts()[0]
font = pg.font.SysFont(f,32)

posP0 = font.render("P0", True, (255, 255, 255), (0, 0, 0))
posP1 = font.render("P1", True, (255, 255, 255), (0, 0, 0))
posP2 = font.render("P2", True, (255, 255, 255), (0, 0, 0))
posP3 = font.render("P3", True, (255, 255, 255), (0, 0, 0))

textRect0 = posP0.get_rect()
textRect1 = posP1.get_rect()
textRect2 = posP2.get_rect()
textRect3 = posP3.get_rect()

P0 = (100,500)
P1 = (200,100)
P2 = (600,80)
P3 = (650,410)

t = 0
speed = 0.004
running = True

while running:
    screen.fill((0,0,0))
    pg.time.delay(50)

    # for event in pg.event.get():
    #     if event.type == pg.QUIT:
    #         running = False

    while(t<1) and running:
        t = t + speed

        BEZ0 = ((1-t)**3 * P0[0], (1-t)**3 * P0[1])
        BEZ1 = (3 * t * (1-t)**2 * P1[0], 3 * t * (1-t)**2 * P1[1])
        BEZ2 = (3 * t * t * (1-t) * P2[0], 3 * t * t * (1-t) * P2[1])
        BEZ3 = ( t**3 * P3[0], t**3  * P3[1])
        x, y = ((BEZ0[0] + BEZ1[0] + BEZ2[0] + BEZ3[0]), (BEZ0[1] + BEZ1[1] + BEZ2[1] + BEZ3[1]))
        textRect0.center = P0 
        textRect1.center = P1
        textRect2.center = P2
        textRect3.center = P3

        screen.blit(posP0, textRect0)
        screen.blit(posP1, textRect1)
        screen.blit(posP2, textRect2)
        screen.blit(posP3, textRect3)
        
        pg.draw.line(screen, (255, 255, 0), P0, P1, 1)
        pg.draw.line(screen, (255, 0, 255), P2, P3, 1)
        pg.draw.circle(screen, (255, 255, 255), (round(x),round(y)), 3)

        pg.display.flip()
    running = False
    
pg.time.delay(1000)
pg.quit()
        
```
