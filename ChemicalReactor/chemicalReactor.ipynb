```python
import matplotlib.pyplot as plt

def main():
    dt = 0.1
    a = 100
    b = 50
    c = 0
    t = 0
    k1 = 0.008
    k2 = 0.002
    N = 60
    A = []
    B = []
    C = []
    A.append(a)
    B.append(b)
    C.append(c)
    print("Initially Time: {}".format(round(t, 2)), "A: {}".format(round(
        a, 2)), "B: {}".format(round(b, 2)), "C: {}".format(round(c, 2)))

    while t<=N:
        a = a + (k2 * c - k1 * a * b) * dt
        b = b + (k2 * c - k1 * a * b) * dt
        c = c + (2 * k1 * a * b - 2 * k2 * c) * dt
        print("At Time: {}".format(round(t, 2)), "A: {}".format(round(
            a, 2)), "B: {}".format(round(b, 2)), "C: {}".format(round(c, 2)))
        A.append(a)
        B.append(b)
        C.append(c)
        t += dt

    plt.plot(A, "b", label = "a")
    plt.plot(B, "r", label = "b")
    plt.plot(C, "g", label = "c")
    plt.legend(title="Legend Title")
    plt.show()
if __name__ == "__main__":
    main()
```
